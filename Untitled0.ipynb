```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import yfinance as yf

# Ensure pandas_datareader uses yfinance backend
import warnings

warnings.filterwarnings("ignore")
# def macd_strategy(symbol):

def macd_strategy(symbol):
  start = dt.datetime(2015, 1, 1)
  end = dt.datetime.now()

  # Fetch gold futures data
  data = yf.download(symbol, start=start, end=end)
  data.columns = data.columns.get_level_values(0)
  day = np.arange(1, len(data) + 1)
  data['day'] = day
  data.drop(columns = ['Volume'], inplace=True)
  data = data[['day', 'Open', 'High' , 'Low', 'Close']]

  data['12-day'] = data['Close'].rolling(12).mean().shift()
  data['26-day'] = data['Close'].rolling(26).mean().shift()

  data['signal'] = np.where(data['12-day'] > data['26-day'], 1, 0)
  data['signal'] = np.where(data['12-day'] < data['26-day'], -1,data['signal'] )
  data.dropna(inplace=True)

  data['return'] = np.log(data['Close']).diff()
  data['system_return'] = data['signal'] * data['return']
  data['entry'] = data.signal.diff()

  buyHold = float(round((np.exp(data['return']).cumprod().iloc[-1] - 1) * 100, 2))
  systemReturn = float(round((np.exp(data['system_return']).cumprod().iloc[-1] - 1) * 100, 2))


  return [symbol ,buyHold, systemReturn, data['Close'].iloc[-1]]

```
